```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.decomposition import TruncatedSVD
from sklearn.linear_model import LogisticRegression
from matplotlib import pyplot
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
file='/content/drive/MyDrive/FINAL PROJEK TEKNIK BIOMEDIKA DAN PARKTIKUM/Program/DATA/80_20 standardisasi filter/test_80_20_standardisasi_filter_8fitur.txt'
dftest = pd.read_csv(file, delimiter='\t')
file2='/content/drive/MyDrive/FINAL PROJEK TEKNIK BIOMEDIKA DAN PARKTIKUM/Program/DATA/80_20 standardisasi filter/training_80_20_standardisasi_filter_8fitur.txt'
dftrain = pd.read_csv(file, delimiter='\t')
```

```python
X_train=dftrain.drop('class',axis=1)
y_train=dftrain['class']

X = dftest.drop('class',axis=1)
y = dftest['class']

X_val, X_test, y_val, y_test = train_test_split(X, y, test_size=0.5, random_state=101)
```

##N=2, 32 detik

```python
svd2 = TruncatedSVD(n_components=2, random_state=101)
X_train=svd2.fit_transform(X_train)
X_val=svd2.fit_transform(X_val)
```

```python
svc=SVC()
#untuk yg n=2, selesai 32 detik
svc.fit(X_train,y_train)
predictions_svd2 = svc.predict(X_val)
print(classification_report(y_val,predictions_svd2))
print(confusion_matrix(y_val,predictions_svd2))
```
